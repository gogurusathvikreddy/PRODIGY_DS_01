```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("PopulationByAgeSex.csv")
x=df['Time']
y=df['PopTotal_0_4']

plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.title('Histogram')
plt.xlabel('Time')
plt.ylabel('total population')
plt.grid(True)

plt.show()
```
